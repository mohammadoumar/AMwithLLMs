```python
def check_nones(x):

    aty_l = x.aty.split(" ")
    flag = 0
    
    for idx, content in enumerate(aty_l):
        if content == 'none' and idx != 0:
            flag = 1
    return flag
```

```python
df_essay['none_check'] = df_essay.apply(lambda x: check_nones(x), axis=1)
```

```python
def get_unique(x):

    input_string = x.rel_pairs

    pairs = re.findall(r'\(.*?\)', input_string)
    #none = re.findall(r'\bnone\b', input_string)
    
    # Combine results and get unique values
    unique_items = set(pairs)
    
    # Convert set to a sorted list if needed (sorting by string representation)
    unique_items = sorted(unique_items)
    
    # Print the result
    return len(unique_items)
```

```python
grouped_df["rel_lengths"] = grouped_df.apply(lambda x: get_unique(x), axis=1)
```
